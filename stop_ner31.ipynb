```python
# Check for unique values in the stop_name column
unique_stops = data_df['stop_name'].unique()
num_unique_stops = len(unique_stops)

# Check if the number of unique stops is reasonable for a NER model
if num_unique_stops > 500:
  print("The number of unique stops is too high for stop_name to be suitable for acting as entities for a NER model.")
else:
  print("The number of unique stops is reasonable for stop_name to be suitable for acting as entities for a NER model.")
```

```python
# prompt: modify the stop name accordingly to fit for ner model

def modify_stop_name(stop_name):
  # Replace spaces with underscores
  stop_name = stop_name.replace(" ", "_")
  # Remove special characters
  stop_name = stop_name.replace(".", "")
  stop_name = stop_name.replace(",", "")
  # Convert to lowercase
  stop_name = stop_name.lower()
  return stop_name

data_df['stop_name'] = data_df['stop_name'].apply(modify_stop_name)
```

```python
# prompt: print the unique stop names working as entities

unique_entities = data_df['stop_name'].unique()

print("Unique stop names working as entities:")
for entity in unique_entities:
  print(f"\t- {entity}")
print(len(unique_entities))
```

```python
# prompt: can these stop names be used as entities to be recognised in a given sentence using NER technique?

# Yes, these stop names can be used as entities to be recognized in a given sentence using NER technique.

# Here's why:

# 1. The number of unique stop names is reasonable (less than 500).
# 2. The stop names have been preprocessed to remove spaces, special characters, and converted to lowercase. This makes them more suitable for NER models, which typically require entities to be in a consistent format.
# 3. The unique stop names have been extracted and stored in a separate list (`unique_entities`). This list can be used to train a NER model to recognize these specific entities in text.

# To use these stop names for NER, you can follow these steps:

# 1. Create a training dataset consisting of sentences that contain the stop names.
# 2. Annotate the stop names in the training dataset with the appropriate entity type (e.g., "STOP_NAME").
# 3. Train a NER model using the training dataset.
# 4. Once the model is trained, you can use it to recognize the stop names in new sentences.
```

```python
text_file = "/content/drive/MyDrive/AIML/Xebia_Internship/data/ner_sentences.txt"
```

```python
# prompt: read the sentence file and check if it can work as training dataset

with open(text_file, 'r') as f:
  sentence_list = f.readlines()

# Check if the file is empty
if not sentence_list:
  print("The sentence file is empty.")
else:
  print("The sentence file is not empty.")

  # Check if the sentences contain any of the unique stop names
  for sentence in sentence_list:
    for entity in unique_entities:
      if entity in sentence:
        print(f"Found entity '{entity}' in sentence: {sentence}")
        break

  # Check if the number of sentences is reasonable for training a NER model
  num_sentences = len(sentence_list)
  if num_sentences < 100:
    print("The number of sentences is too low for training a NER model.")
  else:
    print("The sentence file can be used as a training dataset for a NER model.")
```

```python
# prompt: total number of sentyences in the training dataset

# Get the total number of sentences in the training dataset
total_sentences = len(sentence_list)

# Print the total number of sentences
print(f"Total number of sentences in the training dataset: {total_sentences}")
```

```python
# prompt: Annotate the stop names in the training dataset with the appropriate entity type (e.g., "STOP_NAME").

# Define a function to annotate stop names in a sentence
def annotate_stop_names(sentence):
  entities = []
  for entity in unique_entities:
    if entity in sentence:
      start_index = sentence.index(entity)
      end_index = start_index + len(entity)
      entities.append((start_index, end_index, "STOP_NAME"))
  return entities

# Annotate stop names in all sentences
annotated_sentences = []
for sentence in sentence_list:
  entities = annotate_stop_names(sentence)
  annotated_sentences.append((sentence, entities))

# Print the first 5 annotated sentences
for sentence, entities in annotated_sentences[:50]:
  print(f"Sentence: {sentence}")
  for entity in entities:
    start, end, label = entity
    print(f"\t- Entity: {sentence[start:end]}", f"\t- Label: {label}")
```

```python
# prompt: check if the file is annotated or not

def is_file_annotated(filename):
  with open(filename, 'r') as f:
    for line in f:
      if line.startswith("Sentence:"):
        return True

  return False


# Check if the sentences file is annotated
is_annotated = is_file_annotated(text_file)

# Print the result
if is_annotated:
  print("The sentences file is annotated.")
else:
  print("The sentences file is not annotated.")
```

```python
# prompt: use spacy to annotate the training sentences

import spacy

# Load the spaCy model
nlp = spacy.load("en_core_web_sm")

# Define a function to annotate stop names in a sentence using spaCy
def annotate_stop_names_spacy(sentence):
  doc = nlp(sentence)
  entities = []
  for entity in doc.ents:
    if entity.label_ == "GPE":
      entities.append((entity.start_char, entity.end_char, "STOP_NAME"))
  return entities

# Annotate stop names in all sentences using spaCy
annotated_sentences_spacy = []
for sentence in sentence_list:
  entities = annotate_stop_names_spacy(sentence)
  annotated_sentences_spacy.append((sentence, entities))

# Print the first 5 annotated sentences using spaCy
for sentence, entities in annotated_sentences_spacy[:100]:
  print(f"Sentence: {sentence}")
  for entity in entities:
    start, end, label = entity
    print(f"\t- Entity: {sentence[start:end]}", f"\t- Label: {label}")

```

```python
# prompt: use this file to test for washroom /content/drive/MyDrive/AIML/Xebia_Internship/data/ner_sentences.txt

with open('/content/drive/MyDrive/AIML/Xebia_Internship/data/ner_sentences.txt', 'r') as f:
  sentences = f.readlines()

# Load the trained NER model
#nlp = spacy.load("/content/drive/MyDrive/AIML/Xebia_Internship/stop_ner31")

# Define a function to test the NER model on a new sentence and return the corresponding nearest_washroom_id
def test_ner_and_get_washroom_id(sentence):
  doc = nlp(sentence)
  entities = [(ent.text, ent.label_) for ent in doc.ents]
  if len(entities) == 1 and entities[0][1] == "STOP_NAME":
    stop_name = entities[0][0]
    # Get the nearest_washroom_id for the recognized stop_name from the data_df DataFrame
    nearest_washroom_id = data_df[data_df['stop_name'] == stop_name]['nearest_washroom_id'].values[0]
    return nearest_washroom_id
  else:
    return None

# Test the function on each sentence in the file
for sentence in sentences:
  nearest_washroom_id = test_ner_and_get_washroom_id(sentence)

  # Print the result
  if nearest_washroom_id:
    print(f"The nearest washroom to {sentence.split('to')[1].strip()} is {nearest_washroom_id}.")
  else:
    print("No nearest washroom found for the recognized stop name.")
```
